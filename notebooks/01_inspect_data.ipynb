```python
import pandas as pd
```

```python
import sys
import os

# Get the path to the project root (adjust the number of '..' as needed)
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))

# Add the project root to the system path
if project_root not in sys.path:
    sys.path.append(project_root)
    
```

```python
df = pd.read_csv("../data/raw/btc_1h.csv", parse_dates=["datetime"])
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# 2. Convert the 'timestamp' column to a proper datetime format
# The data downloaded with ccxt is typically in milliseconds
df['datetime'] = pd.to_datetime(df['datetime'], unit='ms')

# 3. Set the 'datetime' column as the index for time-series plotting
df.set_index('datetime', inplace=True)

# 4. Plot the 'close' price
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['close'], label='BTC/USDT Close Price', color='#4285F4')
plt.title('BTC/USDT Closing Price (Last 6 Years)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USDT)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45) # Rotates date labels for better readability
plt.legend()
plt.tight_layout() # Adjusts plot params for tight layout
plt.show()
    
```
